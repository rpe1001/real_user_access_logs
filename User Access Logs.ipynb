```python
import pandas as pd #step 1
df_employees=pd.read_csv('data/employees.csv')
df_system_access=pd.read_csv('data/system_access.csv')
```

```python
df_system_access['id'] = 'EMP' + df_system_access['id'].astype(str).str.zfill(6)
df_system_access.head()
#step2
```

```python
#step 3
df_system_access.rename(columns={'id':'employee_number'},inplace=True)
df_system_access.set_index('employee_number',inplace=True)
df_employees.set_index('employee_number',inplace=True)
df = df_employees.join(df_system_access, how='left')
df
```

```python
df.loc[df['access_level'].isnull()]#step 4
```

```python
subset_df = df.loc[df['access_level'].isnull()].copy()
subset_df #step 5
```

```python
subset_df.loc[subset_df['title'] == 'Director of Sales', 'access_level' ] = 'Approval - Manager'
subset_df.loc[subset_df['title'] == 'Sales Assistant', 'access_level' ] = 'Contract Entity'
subset_df.loc[subset_df['title'] == 'Account Manager', 'access_level' ] = 'Account Admin'
subset_df #step 6

subset_df.to_csv('result.csv') #step 7
```
